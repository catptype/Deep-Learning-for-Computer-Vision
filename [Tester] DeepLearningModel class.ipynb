## Define function with timer

```python
class TimeMeasure:
    @staticmethod
    def execute(model):
        print(f"Running model: {model.func.__name__} ... ", end="")
        try:
            start_time = time.time()
            model()
            end_time = time.time()
            elapsed_time = end_time - start_time
            print(f"PASS {elapsed_time:.4f} seconds.")
        except Exception as e:
            print(f"FAILED ({e})")

    
```
