# Code Testing for Object Oriented Programming in GramCAM

```python
# Installed modules
import matplotlib.pyplot as plt
import tensorflow as tf

# My own modules
import evaluate.GranCAM as GCAM
import script.ImageProcessing as IP
```

## Test case 1: use h5 file then put into GranCAM class

### Create GranCAM model from h5 file

```python
grad_model = GCAM.GranCAM('export model\Res2Net50SE_224x224_2Class (ILLUST VS MANGA).h5')
```

### Prepare image

```python
img_path = 'samples\GranCAM\MANGA.jpg'
img = IP.image_resize_square(img_path, 224, aspect_ratio=True, padding=True)
```

### Generate heatmaps

```python
heatmap = grad_model.generate_heatmap(img, overlay=False)
overlay = grad_model.generate_heatmap(img, overlay=True)

plt.figure(figsize=(10,8))

plt.subplot(1, 2, 1)
plt.imshow(heatmap)
plt.title('Heatmap')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(overlay)
plt.title('Image with heatmap overlay')
plt.axis("off")
```
